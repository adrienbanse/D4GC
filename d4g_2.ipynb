```python
import pandas as pd
```

```python
data = pd.read_csv("statement/compas.csv")
data
```

```python
crits = ['charge_description',
         'priors_count',
         'age',
         'sex',
         'charge_degree',
         'juvenile_felony_count',
         'juvenile_misdemeanors_count',
         'juvenile_other_count',
         'offense_date',
         'screening_date']
sub_data = data[crits]
```

```python
dup = sub_data.duplicated(keep=False)
dup_data = sub_data[dup]
dup_data
```

```python
def compare(i,j):
    l_i = dup_data.loc[i]
    l_j = dup_data.loc[j]
    for crit in crits:
        if l_i[crit]!= l_j[crit]:
            return False
        
    return True
```

```python
for i in dup_data.index :
    print("------------------")
    print(dup_data.loc[i])
    print()
    for j in dup_data.index :
        if compare(i,j):
            print((data.loc[j])[['first_name','last_name','race','score']])
            print()
```
